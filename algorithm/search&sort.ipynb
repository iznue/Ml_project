##### 이진 탐색
- 전체 갯수를 구하고 그 갯수를 1/2로 나눔, 이때 소수점이 나오면 버림
- 1/2 지점의 자료값이 내 검색값보다 큰가 작은가를 비교
- 1/2 지점의 자료값이 내 검색값보다 작으면 우측을 검색 (min 지점을 1/2하고 max 지점은 그대로)
- 1/2 지점의 자료값이 내 검색값보다 크면 좌측을 검색 (min 지점은 그대로 두고, max 지점은 1/2 지점으로 변경)

```python
len(data)

data=list(range(1,12))
print(len(data))

find=8

min,max = 0,len(data)
mid=int(len(data)/2)
mid=int((max-min)/2)
print(min,max,mid)

if data[mid]==find:
    print('data find end')
elif data[mid]<find:
    min=mid
else:
    max=mid

min,max
```

```python
# 기본적으로 while 구문 사용

len(data)

data=list(range(1,12))
print(len(data))

find=8

min,max = 0,len(data)

for x in range(1000):
    mid=int(len(data)/2)
    mid=int((max-min)/2)
    print(min,max,mid)

    if data[mid]==find:
        print('data find end')
        break
    elif data[mid]<find:
        min=mid
        print(min,mid)
        print('찾는 값보다 작음')
    else:
        max=mid

min,max
```

### Sort

```python
# softmax 알고리즘에서 최대값 찾기 -> np.argmax => 2,3 순위 찾아보기

### 최소값, 최대값 찾기
# 최소값 찾기는 첫번째값을 max값으로 지정, 두번째부터 마지막까지 값을 비교하면서 max보다 큰 값이 나오면 max 값을 변경
# -----> 절대 나오지 않을 것 같은 최소 값을 min으로 지정, 두번째부터 마지막까지 값을 비교하면서 min보다 작은값이 나오면 min값 변경

data=[3,9,1,2,8,100,5]
max=9999999
# min=data[0] 메모리에서 꺼내와야 하므로 속도 측면에서 max값을 사용하는 것이 좋음

for i in range(len(data)):
    if max>data[i]:
        max=data[i]

print(max)
# 무조건 첫번째가 최소값이 되도록
```
