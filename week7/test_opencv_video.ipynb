### OpenCV 영상

##### 캠코더 영상 출력

```python
import cv2

webcam = cv2.VideoCapture(0)
webcam.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
webcam.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

if not webcam.isOpened():
    print("Could not open webcam")
    exit()

while webcam.isOpened():
    status, frame = webcam.read()

    if status:
        cv2.imshow("test", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'): # 0또는 1의 경우 계속적으로 읽어옴
        break

webcam.release()
cv2.destroyAllWindows()
```

##### 특정 시간 단위로 화면 저장

```python
import cv2

webcam = cv2.VideoCapture(0)
webcam.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
webcam.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
cnt=0

if not webcam.isOpened():
    print("Could not open webcam")
    exit()

while webcam.isOpened():
    status, frame = webcam.read()
    if status:
        cv2.imshow("test", frame)
        fileName='frame' + str(cnt) + '.jpg'
        cv2.imwrite('d:/output/'+fileName, frame)
        cnt+=1
    if cv2.waitKey(3000) & 0xFF == ord('q'): # cv2.waitKey에 넣는 값에 따라 시간이 지정됨, 3000 : 3초
        break

webcam.release()
cv2.destroyAllWindows()
```

##### 특정 키를 누를 시 화면 저장

```python
import cv2
webcam = cv2.VideoCapture(0)
webcam.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
webcam.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
cnt=0
if not webcam.isOpened():
    print("Could not open webcam")
    exit()
while webcam.isOpened():
    status, frame = webcam.read()
    if status:
        cv2.imshow("test", frame)
 
    if cv2.waitKey(1)==ord('a'):
        fileName='frame' + str(cnt) + '.jpg'
        cv2.imwrite('d:/output/'+fileName, frame)
        print('a키가 눌려서 저장되었습니다.')
        cnt+=1

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

webcam.release()
cv2.destroyAllWindows()
```

##### Video 일정 영역 채우기

```python
import cv2

webcam = cv2.VideoCapture(0)
webcam.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
webcam.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

if not webcam.isOpened():
    print("Could not open webcam")
    exit()

while webcam.isOpened():
    status, frame = webcam.read()
    if status:
        frame[100:300, 200:400]=0
        cv2.imshow("test", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

webcam.release()
cv2.destroyAllWindows()
```
