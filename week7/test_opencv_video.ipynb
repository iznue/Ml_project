### OpenCV 영상

##### 캠코더 영상 출력

```python
import cv2

webcam = cv2.VideoCapture(0)
webcam.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
webcam.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

if not webcam.isOpened():
    print("Could not open webcam")
    exit()

while webcam.isOpened():
    status, frame = webcam.read()

    if status:
        cv2.imshow("test", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'): # 0또는 1의 경우 계속적으로 읽어옴
        break

webcam.release()
cv2.destroyAllWindows()
```

##### 특정 시간 단위로 화면 저장

```python
import cv2

webcam = cv2.VideoCapture(0)
webcam.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
webcam.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
cnt=0

if not webcam.isOpened():
    print("Could not open webcam")
    exit()

while webcam.isOpened():
    status, frame = webcam.read()
    if status:
        cv2.imshow("test", frame)
        fileName='frame' + str(cnt) + '.jpg'
        cv2.imwrite('d:/output/'+fileName, frame)
        cnt+=1
    if cv2.waitKey(3000) & 0xFF == ord('q'): # cv2.waitKey에 넣는 값에 따라 시간이 지정됨, 3000 : 3초
        break

webcam.release()
cv2.destroyAllWindows()
```

##### 특정 키를 누를 시 화면 저장

```python
import cv2
webcam = cv2.VideoCapture(0)
webcam.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
webcam.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
cnt=0
if not webcam.isOpened():
    print("Could not open webcam")
    exit()
while webcam.isOpened():
    status, frame = webcam.read()
    if status:
        cv2.imshow("test", frame)
 
    if cv2.waitKey(1)==ord('a'):
        fileName='frame' + str(cnt) + '.jpg'
        cv2.imwrite('d:/output/'+fileName, frame)
        print('a키가 눌려서 저장되었습니다.')
        cnt+=1

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

webcam.release()
cv2.destroyAllWindows()
```

### video 파일에서 얼굴 영역 검출

```python
from keras.models import load_model
model = load_model('keras_model.h5')  

#model.weights

import cv2
import numpy as np
class_name=['Female','Male']     
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

webcam = cv2.VideoCapture(0)
webcam.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
webcam.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

if not webcam.isOpened():
    print("Could not open webcam")
    exit()

while webcam.isOpened():
    status, frame = webcam.read()

    if status:
        cv2.imshow("test", frame)
        
        # 얼굴 찾기
        faces = face_cascade.detectMultiScale(frame, 1.3, 5)

        cnt=0
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x-20, y-20), (x + w+20, y + h+20), (255, 0, 0), 2)
            사각영역array = frame[y-20:y + h+20, x-20:x + w+20]
            print(x,y,w,h)
            
            ### 모델에 넣기 위한 작업
            resize_img = cv2.resize(사각영역array, (224, 224))
            imgArr=(resize_img/127.0)-1   # 티처블 코드에 있었음
            prediction=imgArr.reshape(1,224,224,3) # 인공지능은  array 를 하나 확장한 tensor로 해야함. (224,224,3)을 1,224,224,3 으로 확장함
            result=model.predict(prediction)
            argmaxNumber=np.argmax(result)   # 출력결과가 0이면 0번째 위치값이 predict 출력값중에 0번째 위치값이 가장 크다는 뜻
            printTxt=class_name[argmaxNumber]
            cv2.putText(frame, printTxt , (x-10,y-30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2)
            cv2.imshow("test", frame)



    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

webcam.release()
cv2.destroyAllWindows()
```
