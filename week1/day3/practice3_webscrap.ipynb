```python
import sys
import time
from selenium import webdriver
from selenium.webdriver import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup


# 크롬 웹드라이버 설치 및 설정
chrome_options = webdriver.ChromeOptions()

# 영상을 멈춘 후 스크롤 및 페이지다운
# selenium을 이용하여 액션을 주며 데이터를 가져옴
```

```python
# user로 보이기 위한 user_agent -> 각자의 컴퓨터에 맞게 확인이 필요함
# bot이 아니라는 것을 보여주기 위한 정보가 필요
# 구글링 user agent string 검색 후 내 노트북에 맞는 정보를 가져옴
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'
chrome_options.add_argument('user-agent='+user_agent)
```

```python
chrome_options.add_experimental_option('detach', True)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=chrome_options) # driver를 install함
driver.maximize_window() # 윈도우 창 확대

# 설정
keyword = 'Me at the zoo' # 검색할 제목
count = 10 # 스크래핑할 댓글 개수 (값을 -1로 하면 한계까지 수집됨)

# 크롬 열기 및 주소입력
driver.get('https://www.youtube.com/results?search_query='+keyword)
driver.implicitly_wait(5) # wait를 주고 time sleep을 하는 것이 좋음
time.sleep(1)

# 영상 선택
# //*[@id="thumbnail"]/yt-image/img -> copy xpath
driver.find_element(By.XPATH, '//*[@id="thumbnail"]/yt-image/img').click()  # f12 key -> click하여 원하는 정보를 얻음
# html 요소를 가져와서 선택함 
driver.implicitly_wait(5) # wait를 주고 time sleep을 하는 것이 좋음
time.sleep(1)

# 재생중인 영상 중지
driver.find_element(By.XPATH, '//*[@id="movie_player"]/div[30]/div[2]/div[1]/button').click()

driver.find_element(By.CSS_SELECTOR, 'body').send_keys(Keys.END)
time.sleep(1)


# 필요한 댓글 확보를 위한 while문
# 스크롤 높이를 비교하여 더 내릴지 결정
while True:
    # 스크롤 전 현재높이 저장
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    print('new_height : ',new_height)
    time.sleep(1) #1초 대기

    # while문 멈출지 체크
    if count == -1:  # 페이지 끝까지 while문 돌리기, 얼마나 스크롤 할지 모를 때 임시로 -1 지정
        # 페이지 끝까지 스크롤 내리기
        driver.find_element(By.CSS_SELECTOR, 'body').send_keys(Keys.END)
        time.sleep(1)

        # 스크롤 후 현재높이 저장
        last_height = driver.execute_script("return document.documentElement.scrollHeight")
        print('last_height : ', last_height) # 스크롤 한 높이 측정
        time.sleep(1)

        # 페이지 끝까지 스크롤 했는지 판단
        if new_height == last_height:  # if(이전 스크롤 높이 = 현재 스크롤 높이) -> 마지막 페이지라 판단하는 경우 while문 탈출
            break
    else:
        # 페이지 읽기
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        time.sleep(1)

        # 댓글 구역 추출
        comments_sections = soup.find_all('ytd-comment-thread-renderer', class_='ytd-item-section-renderer')  # html source code, ()안 : 댓글 부분
        
        if len(comments_sections) < count:  # 수집하려는 개수보다 확보된 댓글이 많다면
            # 페이지 끝까지 스크롤 내리기
            driver.find_element(By.CSS_SELECTOR, 'body').send_keys(Keys.END)
            time.sleep(1)

            # 스크롤 후 현재높이 저장
            last_height = driver.execute_script("return document.documentElement.scrollHeight")
            print('last_height : ', last_height) # 스크롤 한 높이 측정
            time.sleep(1)

            # 페이지 끝까지 스크롤 했는지 판단
            if new_height == last_height:  # if(이전 스크롤 높이 = 현재 스크롤 높이) -> 마지막 페이지라 판단하는 경우 while문 탈출
                break
        else:
            print(f'{count}만큼 댓글 확보 완료')
            del comments_sections[count:] # 댓글 구역에서 count보다 초과된 댓글 삭제
            # comments_sections에 review에 해당되는 내용들이 들어감
            break

print('페이지 스크롤 완료')
```

```python
username = []
usercomment = []
# username과 usercomment외에 다른 요소는 어떻게 가져올지? 

for item in comments_sections:
    try:
        a = item.find('yt-formatted-string', id='text').get_text()
        b = item.find('yt-formatted-string', id='content-text').get_text()
    except:
        a = item.find('a', id='author-text').get_text().strip()
        b = item.find('yt-formatted-string', id='content-text').get_text()
    username.append(a)
    usercomment.append(b)

# 확인
print(f'count: {len(usercomment)}')
print('{username: usercomment} is')
for i, j in zip(username, usercomment):
    print(f'{{{i}: {j}}}')

# 다음 정보를 csv 파일로 저장해보기
```
